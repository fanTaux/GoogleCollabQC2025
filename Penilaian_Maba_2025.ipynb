```python
import pandas as pd
from google.colab import files
import io

print("Silakan upload 2 file CSV Anda.")
print("1. File daftar mahasiswa (berisi kolom 'NIM' dan 'Nama').")
print("2. File nilai penugasan (berisi kolom 'NIM', 'Nama', 'Nilai', dll).")

try:
    uploaded = files.upload()

    if len(uploaded) != 2:
        print("\n⚠️ Harap upload tepat 2 file CSV.")
    else:
        file_names = list(uploaded.keys())
        file1_name = file_names[0]
        file2_name = file_names[1]

        print(f"\nFile '{file1_name}' dan '{file2_name}' berhasil di-upload.")

        try:
            df1_test = pd.read_csv(io.BytesIO(uploaded[file1_name]))
            df2_test = pd.read_csv(io.BytesIO(uploaded[file2_name]))

            if len(df1_test.columns) < len(df2_test.columns):
                df_mahasiswa = df1_test
                df_nilai = df2_test
                print("\nFile daftar mahasiswa terdeteksi sebagai:", file1_name)
                print("File nilai penugasan terdeteksi sebagai:", file2_name)
            else:
                df_mahasiswa = df2_test
                df_nilai = df1_test
                print("\nFile daftar mahasiswa terdeteksi sebagai:", file2_name)
                print("File nilai penugasan terdeteksi sebagai:", file1_name)


            if 'NIM' not in df_mahasiswa.columns or 'Nama' not in df_mahasiswa.columns:
                raise ValueError("File daftar mahasiswa harus memiliki kolom 'NIM' dan 'Nama'")
            if 'NIM' not in df_nilai.columns or 'Nilai' not in df_nilai.columns:
                raise ValueError("File nilai penugasan harus memiliki kolom 'NIM' dan 'Nilai'")


            df_nilai_selected = df_nilai[['NIM', 'Nilai']]

            hasil_df = pd.merge(df_mahasiswa, df_nilai_selected, on='NIM', how='left')

            # Corrected inplace operation
            hasil_df['Nilai'] = hasil_df['Nilai'].fillna(0)

            hasil_df['Nilai'] = hasil_df['Nilai'].astype(int)


            print("\n✅ Proses perbandingan dan pengolahan nilai selesai.")
            print("\nPreview hasil akhir:")
            print(hasil_df.head())

            output_filename = 'hasil_nilai_maba.xlsx'
            hasil_df.to_excel(output_filename, index=False)


            print(f"\nFile output '{output_filename}' telah dibuat dan siap diunduh.")
            files.download(output_filename)

        except Exception as e:
            print(f"\nTerjadi kesalahan saat membaca file CSV: {e}")
            print("Pastikan file yang diupload berformat CSV dan memiliki kolom yang sesuai.")


except Exception as e:
    print(f"\nTerjadi kesalahan saat proses upload: {e}")
```
