## Untuk Nilai Streak Simaba

```python
import pandas as pd
from google.colab import files
import io

print("Silakan upload 2 file Anda.")
print("Sistem akan otomatis membersihkan nama kolom (menghapus spasi & kapitalisasi).")

try:
    uploaded = files.upload()

    if len(uploaded) != 2:
        print("\n‚ö†Ô∏è Harap upload tepat 2 file.")
    else:
        file_names = list(uploaded.keys())

        def read_file(file_name, uploaded_content):
            file_extension = file_name.split('.')[-1].lower()
            if file_extension == 'csv':
                # Mencoba membaca CSV, jika separatornya titik koma (;) akan dideteksi otomatis atau default koma
                try:
                    return pd.read_csv(io.BytesIO(uploaded_content[file_name]))
                except:
                    return pd.read_csv(io.BytesIO(uploaded_content[file_name]), sep=';')
            elif file_extension == 'xlsx':
                return pd.read_excel(io.BytesIO(uploaded_content[file_name]))
            else:
                raise ValueError(f"Format file {file_name} tidak didukung.")

        dfs = []

        # --- TAHAP 1: BACA DAN BERSIHKAN KOLOM ---
        for name in file_names:
            try:
                raw_df = read_file(name, uploaded)

                # PENTING: Membersihkan nama kolom (Hapus spasi & jadikan Huruf Besar semua)
                # Contoh: "  nim  " menjadi "NIM", "Nilai Tugas" menjadi "NILAI TUGAS"
                raw_df.columns = [str(col).strip().upper() for col in raw_df.columns]

                print(f"\nüìÑ Membaca file: {name}")
                print(f"   Kolom yang ditemukan: {list(raw_df.columns)}") # Debugging print

                dfs.append({'name': name, 'df': raw_df})
            except Exception as e:
                print(f"Gagal membaca file {name}: {e}")

        # --- TAHAP 2: IDENTIFIKASI FILE ---
        df_mahasiswa = None
        df_nilai = None

        for item in dfs:
            cols = item['df'].columns

            # Cek keberadaan kolom wajib (sudah uppercase semua)
            if 'NILAI' in cols:
                df_nilai = item['df']
                print(f"   -> Terdeteksi sebagai FILE NILAI")
            elif 'NIM' in cols and 'NAMA' in cols:
                df_mahasiswa = item['df']
                print(f"   -> Terdeteksi sebagai FILE MAHASISWA")
            elif 'NIM' in cols:
                # Jika punya NIM tapi tidak punya Nilai, asumsi file mahasiswa
                df_mahasiswa = item['df']
                print(f"   -> Terdeteksi sebagai FILE MAHASISWA (Berdasarkan NIM)")

        # --- TAHAP 3: VALIDASI & FILTER ---
        if df_mahasiswa is None:
            raise ValueError("‚ùå Gagal menemukan File Mahasiswa (Wajib ada kolom 'NIM' dan 'NAMA')")
        if df_nilai is None:
            raise ValueError("‚ùå Gagal menemukan File Nilai (Wajib ada kolom 'NIM' dan 'NILAI')")

        # Ambil hanya NIM dan NILAI saja
        # Karena kita sudah uppercase semua kolom di awal, kita panggil dengan 'NIM' dan 'NILAI'
        df_nilai_selected = df_nilai[['NIM', 'NILAI']].copy()

        # --- TAHAP 4: MERGE ---
        # Konversi NIM ke string agar tipe data sama (menghindari error beda format text vs number)
        df_mahasiswa['NIM'] = df_mahasiswa['NIM'].astype(str).str.strip()
        df_nilai_selected['NIM'] = df_nilai_selected['NIM'].astype(str).str.strip()

        hasil_df = pd.merge(df_mahasiswa, df_nilai_selected, on='NIM', how='left')

        # Rapikan Nilai
        hasil_df['NILAI'] = hasil_df['NILAI'].fillna(0)
        try:
            hasil_df['NILAI'] = hasil_df['NILAI'].astype(int)
        except:
            pass # Jika gagal jadi int (misal ada huruf), biarkan apa adanya

        # --- TAHAP 5: OUTPUT ---
        print("\n‚úÖ Proses Berhasil!")
        print(hasil_df.head())

        output_filename = 'hasil_rekap_nilai_fix.xlsx'
        hasil_df.to_excel(output_filename, index=False)
        print(f"\nFile '{output_filename}' siap diunduh.")
        files.download(output_filename)

except Exception as e:
    print(f"\n‚ùå ERROR: {e}")
    print("Saran: Cek 'Kolom yang ditemukan' di atas. Apakah NIM terbaca sebagai 'UNNAMED: 0'?")
    print("Jika ya, berarti header tabel Anda tidak di baris pertama Excel.")
```

## Untuk Email Maba

```python
import pandas as pd
from google.colab import files
import io

print("Silakan upload 2 file Anda (CSV atau XLSX).")
print("1. File Acuan Urutan (berisi kolom 'NIM' dan 'Nama').")
print("2. File Data Lengkap (berisi kolom 'nim', 'name', 'prodi', 'cluster', 'group', 'email').")

try:
    uploaded = files.upload()

    if len(uploaded) != 2:
        print("\n‚ö†Ô∏è Harap upload tepat 2 file (CSV atau XLSX).")
    else:
        file_names = list(uploaded.keys())
        file1_name = file_names[0]
        file2_name = file_names[1]

        print(f"\nFile '{file1_name}' dan '{file2_name}' berhasil di-upload.")

        def read_file(file_name, uploaded_files):
            file_extension = file_name.split('.')[-1].lower()
            if file_extension == 'csv':
                return pd.read_csv(io.BytesIO(uploaded_files[file_name]))
            elif file_extension == 'xlsx':
                return pd.read_excel(io.BytesIO(uploaded_files[file_name]))
            else:
                raise ValueError(f"Format file tidak didukung: {file_extension}. Harap upload file CSV atau XLSX.")

        try:
            # Baca kedua file
            df_a = read_file(file1_name, uploaded)
            df_b = read_file(file2_name, uploaded)

            # --- LOGIKA PENENTUAN FILE ---
            # Kita cek file mana yang punya kolom 'email' (case insensitive)
            # File yang punya email adalah File Kedua (sumber data)
            # File sisanya adalah File Pertama (acuan urutan)

            # Normalisasi nama kolom sementara untuk pengecekan (ubah ke lowercase)
            cols_a = [c.lower() for c in df_a.columns]
            cols_b = [c.lower() for c in df_b.columns]

            if 'email' in cols_a:
                df_data_lengkap = df_a # File Kedua
                df_acuan = df_b        # File Pertama
                print(f"\nFile Data Lengkap (Sumber Email) terdeteksi: {file1_name}")
                print(f"File Acuan Urutan terdeteksi: {file2_name}")
            elif 'email' in cols_b:
                df_data_lengkap = df_b # File Kedua
                df_acuan = df_a        # File Pertama
                print(f"\nFile Data Lengkap (Sumber Email) terdeteksi: {file2_name}")
                print(f"File Acuan Urutan terdeteksi: {file1_name}")
            else:
                raise ValueError("Tidak ditemukan kolom 'email' di kedua file. Pastikan format file sesuai.")

            # --- PEMBERSIHAN NAMA KOLOM ---
            # Pastikan kolom NIM konsisten (Ubah 'nim' menjadi 'NIM' agar bisa di-merge)
            df_acuan.columns = df_acuan.columns.str.replace('nim', 'NIM', case=False)
            df_data_lengkap.columns = df_data_lengkap.columns.str.replace('nim', 'NIM', case=False)
            df_data_lengkap.columns = df_data_lengkap.columns.str.replace('email', 'Email', case=False)

            # Validasi kolom wajib
            if 'NIM' not in df_acuan.columns:
                raise ValueError("File Acuan harus memiliki kolom 'NIM' atau 'nim'.")
            if 'NIM' not in df_data_lengkap.columns or 'Email' not in df_data_lengkap.columns:
                raise ValueError("File Data Lengkap harus memiliki kolom 'nim' dan 'email'.")

            # --- PENGGABUNGAN DATA (MERGE) ---
            # Ambil hanya NIM dan Email dari data lengkap
            df_sumber_email = df_data_lengkap[['NIM', 'Email']]

            # Merge menggunakan how='left' agar urutan mengikuti df_acuan (File Pertama)
            # dan baris tidak hilang meskipun tidak ada emailnya.
            hasil_df = pd.merge(df_acuan, df_sumber_email, on='NIM', how='left')

            # --- OUTPUT ---
            # Hanya ambil kolom NIM dan Email sesuai permintaan
            output_df = hasil_df[['NIM', 'Email']]

            print("\n‚úÖ Proses pencocokan selesai.")
            print("\nPreview hasil akhir (sesuai urutan File Pertama):")
            print(output_df.head())

            output_filename = 'hasil_nim_email.xlsx'
            output_df.to_excel(output_filename, index=False)

            print(f"\nFile output '{output_filename}' telah dibuat dan siap diunduh.")
            files.download(output_filename)

        except Exception as e:
            print(f"\nTerjadi kesalahan saat memproses data: {e}")

except Exception as e:
    print(f"\nTerjadi kesalahan saat proses upload: {e}")
```
